# Análise da Folha de Pagamento — Prefeitura de Santa Rita do Passa Quatro/SP (2025)

## 01 — Business Understanding (CRISP-DM)

Este projeto tem como objetivo analisar e transformar os dados da folha de pagamento dos servidores públicos da Prefeitura de **Santa Rita do Passa Quatro/SP**, referentes ao ano de **2025**, em informações claras, úteis e acionáveis.

Para garantir estrutura, reprodutibilidade e rigor metodológico, o trabalho segue a metodologia **CRISP-DM (Cross-Industry Standard Process for Data Mining)**, amplamente adotada em projetos profissionais de Ciência de Dados.

Nesta etapa inicial, definimos o **contexto**, os **objetivos de negócio**, as **questões investigativas** e os **critérios que orientarão as próximas fases** do projeto.

## Contexto do Negócio

A disponibilização de dados públicos por meio do **Portal da Transparência** permite avaliar como os recursos municipais são aplicados, especialmente no que diz respeito às despesas com **pessoal**, que representam uma das maiores parcelas do orçamento público.

Fonte dos dados:  
- Portal da Transparência — Prefeitura de Santa Rita do Passa Quatro/SP  
  https://www.transparencia.prefsrpq.com.br/transparencia/

A análise estruturada dessa base possibilita compreender a **composição da folha de pagamento**, identificar **padrões de remuneração**, avaliar **distribuições por categoria funcional** e detectar eventuais **concentrações ou distorções** que possam indicar oportunidades de melhoria na gestão pública.



## Objetivos do Projeto

### Objetivo Geral
Realizar uma análise exploratória estruturada da folha de pagamento de **2025**, transformando os dados em **insights relevantes** por meio de estatística descritiva, visualização analítica e modelagem básica, garantindo rigor metodológico e reprodutibilidade.

### Objetivos Específicos
- Mapear a **distribuição salarial** entre servidores e categorias funcionais.
- Avaliar diferenças de remuneração entre **áreas**, **secretarias** e **tipos de vínculo**.
- Investigar possíveis assimetrias relacionadas ao **gênero inferido**, categorias e níveis hierárquicos.
- Analisar padrões temporais ao longo do ano (variações mensais de remuneração e composição do quadro).
- Identificar o impacto de **cargos comissionados** no gasto total.
- Preparar uma base tratada e documentada, garantindo **anonimização** e adequação às boas práticas de transparência de dados públicos.
- Viabilizar etapas posteriores de modelagem ou construção de dashboards.

## Perguntas Analíticas

As questões abaixo orientam a exploração inicial dos dados e direcionam a geração de insights na etapa de EDA:

- Como se distribuem os salários dos servidores públicos municipais em 2025?
- Quais cargos, categorias ou secretarias concentram o maior custo total de pessoal?
- Existem diferenças salariais relevantes entre áreas, níveis hierárquicos ou vínculos (efetivos, comissionados etc.)?
- Há variações ou padrões temporais significativos ao longo do ano (sazonalidade ou mudanças estruturais)?
- Existem **outliers** salariais que demandam investigação mais detalhada?
- A análise permite identificar agrupamentos naturais (clusters) de servidores com perfis semelhantes de remuneração?

## Considerações Éticas e LGPD

Embora os dados utilizados sejam **públicos** e disponibilizados oficialmente pelo Portal da Transparência, este projeto adota boas práticas de **privacidade, ética e conformidade com a LGPD**, incluindo:

- Anonimização de nomes e qualquer informação que possa identificar indivíduos.
- Não exposição de dados sensíveis ou inferências que possam causar constrangimento ou interpretação indevida.
- Uso dos dados exclusivamente para fins **educacionais**, **analíticos** e de **transparência pública**, sem finalidade comercial.
- Documentação clara das transformações aplicadas, garantindo rastreabilidade e respeito ao caráter público dos dados.
