# Análise da Folha de Pagamento – Prefeitura de Santa Rita do Passa Quatro/SP (2025)

## Business Understanding (CRISP-DM)

Este projeto tem como objetivo realizar uma análise exploratória e modelagem
dos dados da folha de pagamento dos servidores públicos da Prefeitura de
Santa Rita do Passa Quatro – SP, referentes ao ano de 2025.

A análise segue rigorosamente a metodologia **CRISP-DM**, amplamente utilizada
em projetos profissionais de Ciência de Dados.


## Contexto do Negócio

A disponibilização de dados públicos por meio do Portal da Transparência
permite analisar como os recursos públicos são utilizados, especialmente
no que se refere aos gastos com pessoal.

Link do Portal da Transparência:
- https://www.transparencia.prefsrpq.com.br/transparencia/

A análise desses dados possibilita compreender a estrutura de remuneração
dos servidores públicos municipais, identificar padrões, assim como
potenciais distorções ou concentrações atípicas.


## Objetivos do Projeto

### Objetivo Geral
Realizar uma análise exploratória completa da folha de pagamento de 2025,
extraindo insights relevantes por meio de estatística descritiva,
visualização de dados e técnicas de modelagem.

### Objetivos Específicos
- Analisar a distribuição dos salários dos servidores
- Avaliar diferenças de remuneração entre cargos e secretarias
- Investigar padrões temporais ao longo do ano
- Aplicar técnicas estatísticas e/ou de Machine Learning
- Garantir anonimização dos dados pessoais


## Perguntas Analíticas

- Como se distribuem os salários dos servidores públicos?
- Quais cargos concentram os maiores gastos com pessoal?
- Existem diferenças salariais relevantes entre secretarias?
- Há evidências de sazonalidade ao longo do ano?
- Existem outliers salariais que merecem investigação?
- É possível agrupar servidores com perfis salariais semelhantes?


## Considerações Éticas e LGPD

Apesar de os dados serem públicos, este projeto adota boas práticas de
privacidade e ética, incluindo:

- Anonimização dos nomes dos servidores públicos
- Não exposição de informações sensíveis
- Uso dos dados exclusivamente para fins educacionais e analíticos
